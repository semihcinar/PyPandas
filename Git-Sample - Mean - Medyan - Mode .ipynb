```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
samples = np.random.normal(loc= 100, scale = 15,size=1000)
samples
```

## Compute the **mean**, **median**, and **mode**

```python
mean = np.mean(samples)
median = np.median(samples)
mode = stats.mode(samples)
```
